# Univariate Analysis (catagorical)

```python
sns.countplot(data=df, x=df['Survived'])
```

```python
(df['Survived'].value_counts() / len(df) ) * 100
```

```python
survived_count = df['Survived'].value_counts()

survived_level = df['Survived'].unique()

plt.pie(survived_count, labels=survived_level, autopct='%1.1f')
plt.title('Pie Chart of Servived Rate')
plt.show()
```

# Univariate Analysis (Numerical)

```python
sns.histplot(df['Age'], bins=50)
plt.title('Histrgram of Age')
plt.xlabel('Ages')
plt.ylabel('Frequency')
```

```python
sns.kdeplot(df['Age'])
plt.title('KDE plot of Age')
plt.xlabel('Ages')
plt.ylabel('Probality Distribution')
plt.show()
```

```python
sns.histplot(df['Fare'], bins=50)
plt.title('Histrgram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
```

```python
sns.kdeplot(df['Fare'])
plt.title('Histrgram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
```
