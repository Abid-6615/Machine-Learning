```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_updated.csv')
```

```python
df.sample(10)
```

# Multi Variate Analysis

```python
sns.countplot(x=df['Sex'], hue=df['Survived'])
```

```python
gender_split = df.groupby('Sex')['Survived'].value_counts(normalize=True)

print(gender_split)
```

```python
sns.countplot(x=df['Pclass'], hue=df['Survived'])
```

pclass_split = df.groupby('Pclass')['Survived'].value_counts(normalize=True)

print(pclass_split)

# Multi Variate Analysis (Barplot for Cat-Num)

```python
sns.barplot(x=df['Pclass'], y=df['Fare'])
```

```python
sns.barplot(x=df['Survived'], y=df['Age'], hue=df['Sex'])
```

# Multivariate Analysis (KDE plot for categorical discrepancy)


```python
sns.kdeplot(data=df, x=df['Age'], hue='Survived')
```
